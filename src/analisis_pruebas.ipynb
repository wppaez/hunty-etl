```python
import sqlite3
import pandas as pd

# Load the Excel file into a DataFrame
excel_file_path = 'C:/Users/wppaez/Desktop/hunty-etl/output/responses.xlsx'
df = pd.read_excel(excel_file_path)

# Connect to SQLite database (create one if it doesn't exist)
sqlite_conn = sqlite3.connect('my_database.db')

# Write the DataFrame to SQLite database
df.to_sql('responses_table', sqlite_conn, if_exists='replace', index=False)

# Commit changes and close the connection
sqlite_conn.commit()
sqlite_conn.close()

print("Excel data uploaded to SQLite database.")
```

```python
import sqlite3
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('my_database.db')

# Ejecutar la consulta SQL para contar los IDs diferentes de 'NULL'
query = """
    SELECT COUNT(*) AS non_null_ids
    FROM users_table
    WHERE ID <> 'NULL'
"""
result = conn.execute(query)
non_null_ids_count = result.fetchone()[0]

# Calcular el porcentaje
total_ids_query = "SELECT COUNT(*) AS total_ids FROM users_table"
total_result = conn.execute(total_ids_query)
total_ids = total_result.fetchone()[0]
percentage = (non_null_ids_count / total_ids) * 100

# Cerrar la conexión a la base de datos
conn.close()

# Crear la gráfica de barras amarilla
labels = ['Con nombre completo', 'Sin nombre completo']
sizes = [percentage, 100 - percentage]
colors = ['#FFD700', '#FFA07A']  # Amarillo para IDs con valor, salmón para IDs con NULL

plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=colors)
plt.xlabel('Categorías')
plt.ylabel('Porcentaje')
plt.title('Porcentaje de usuarios con NOMBRE COMPLETO')
plt.ylim(0, 100)  # Asegurar que el eje y vaya de 0 a 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mejor visualización
plt.tight_layout()  # Ajustar diseño de la gráfica
plt.show()

```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('my_database.db')

# Ejecutar la consulta SQL para contar las conversaciones con la condición especificada
query_condition = """
    SELECT COUNT(*) AS conversaciones_condicion
    FROM conversaciones
    WHERE [ID Pregunta] = '2' AND Respuesta = '[CIUDAD]'
"""
result_condition = pd.read_sql_query(query_condition, conn)
conversaciones_condicion = result_condition['conversaciones_condicion'].iloc[0]

# Calcular el total de conversaciones
query_total = "SELECT COUNT(distinct [Conversación ID]) AS total_conversaciones FROM conversaciones"
result_total = pd.read_sql_query(query_total, conn)
total_conversaciones = result_total['total_conversaciones'].iloc[0]

# Calcular el porcentaje
if total_conversaciones != 0:
    porcentaje_condicion = (conversaciones_condicion / total_conversaciones) * 100
else:
    porcentaje_condicion = 0

# Crear la gráfica de torta con colores morados
labels = ['Respondieron', 'No Respondieron']
sizes = [porcentaje_condicion, 100 - porcentaje_condicion]
colors = ['#8A2BE2', '#BA55D3']  # Morado claro y morado oscuro

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Porcentaje que respondieron pregunta 2')
plt.axis('equal')  # asegurar que la gráfica sea un círculo
plt.show()

# Cerrar la conexión a la base de datos
conn.close()

```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('my_database.db')

# Ejecutar la consulta SQL para contar las conversaciones con la condición especificada
query_condition = """
    SELECT COUNT(*) AS conversaciones_condicion
    FROM conversaciones
    WHERE [ID Pregunta] = '3' AND Respuesta <> 'NULL'
"""
result_condition = pd.read_sql_query(query_condition, conn)
conversaciones_condicion = result_condition['conversaciones_condicion'].iloc[0]

# Calcular el total de conversaciones
query_total = "SELECT COUNT(distinct [Conversación ID]) AS total_conversaciones FROM conversaciones"
result_total = pd.read_sql_query(query_total, conn)
total_conversaciones = result_total['total_conversaciones'].iloc[0]

# Calcular el porcentaje
if total_conversaciones != 0:
    porcentaje_condicion = (conversaciones_condicion / total_conversaciones) * 100
else:
    porcentaje_condicion = 0

# Crear la gráfica de barras
labels = ['Respondieron', 'No Respondieron']
sizes = [porcentaje_condicion, 100 - porcentaje_condicion]
colors = ['#5cb85c', '#d9534f']  # Verde para respondieron, rojo para no respondieron

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, sizes, color=colors)
plt.xlabel('Respuestas')
plt.ylabel('Porcentaje')
plt.title('Porcentaje que respondieron pregunta 3')
plt.ylim(0, 100)  # Asegurar que el eje y vaya de 0 a 100
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el porcentaje en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval - 5, f'{yval:.1f}%', va='center', ha='center', color='white', fontsize=10)

plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mejor visualización
plt.tight_layout()  # Ajustar diseño de la gráfica
plt.show()

# Cerrar la conexión a la base de datos
conn.close()
```
